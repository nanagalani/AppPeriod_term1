```python
f = open('whomwho.txt', 'r')
whomwho = {}
for line in f.readlines():
    if line.split(',')[0] in whomwho.keys():
        whomwho[line.split(',')[0]].append(line.split(',')[1][:-1])
    else:
        whomwho[line.split(',')[0]] = [line.split(',')[1][:-1]]
```

```python
import numpy as np
unique_infl = []
unique_infl = np.unique(whomwho.keys()).tolist()
```

```python
n_infl = len(unique_infl)
infl_matrix = np.zeros(shape=(n_infl,n_infl))
for whom in whomwho.keys():
    for who in whomwho[whom]:
        if who in unique_infl:
            infl_matrix[unique_infl.index(whom)][unique_infl.index(who)] = 1
```

```python
all_fol = []
for val, val2 in whomwho.items():
    for fol in val2:
        all_fol.append(fol);
print len(all_fol)
unique_fol = np.unique(all_fol)
```

```python
import numpy as np
import networkx as nx

#Make it column stohastic
def normAdjMatrix(A):
    n = np.size(A,1)
    #diagonalize an adjacency matrix
    for i in xrange(n): A[i,i] = 1
        
    A = A.astype(np.float)
    for col in A.T[:,]:
        if sum(col) != 0:
            col /= sum(col)
    return A
```

```python
def PRPower(A, acc = 1e-6):
    iter = 0
    maxIter = 100
    rank = np.ones(A.shape[0])/A.shape[0]
    while iter < maxIter:
        newrank = A.dot(rank)
        iter += 1
        if np.linalg.norm(newrank - rank) < abs(acc):
            break
        rank = newrank
       
    return (rank, iter)

def PREigen(A):
    w, v = np.linalg.eig(A)
    w = np.round(w, decimals=10)
    #print w
    v = np.round(v, decimals=10)
    rank = np.real(v[:, np.where(w == 1)[0]])
    rank /= sum(rank)
    return rank.T

A = normAdjMatrix(infl_matrix)

#start = time.time()
power, iters = PRPower(A)
#t = time.time() - start

#start = time.time()
eigen = PREigen(A)
print np.shape(power)
print np.shape(eigen)
#t = time.time() - start
```

```python
#HITS: an authority value a_i and a hub value h_i. 
#Authority values are simply the sum of the in-neighbor hub values, and similarly the hub values are the sum of the out-neighbor authority values. 
#In matrix form we have the following two matrix-vector equations: a= A’h and h = Aa
#Plugging back in for h and a leads to the following equations yielding an pair of eigenvector problems: a= A’Aa  and  h = AA’h
import numpy as np
def HITS(A):
    ksteps = 5
    n = len(A)
    Au = np.dot(np.transpose(A),A)
    Hu = np.dot(A,np.transpose(A))
    a = np.ones(n); 
    h = np.ones(n)
    #print a,h
    for j in range(ksteps):
        a = np.dot(a,Au)
        a= a/sum(a)
        h = np.dot(h,Hu)
        h = h/ sum(h)
        #print a,'\n',h
    return (a,h)
```
